Linear Regression From Scatch per il libro How Machines Learn By Roberto

```python
import numpy as np 

# Anni di esperienza (array di numeri)
X = np.array([24,14,5,15,26,35,8,0,1,16,17,37,4,6])
# Stipendio basato su anni di esperienza (inventato)
y = np.array([2880,2000,1700,2230,2993,3012,1892,1100,1231,2103,2189,3231,1620,2018])
```

```python
import matplotlib.pyplot as plt

plt.scatter(X, y)
plt.show()
```

```python
m = 0 # Coefficiente angolare (w)
q = 0 # Intercetta
y_pred = q + m * X
y_pred
```

```python
def derivatives(x, y, y_pred):
    dq = np.mean(y_pred - y)
    dm = np.mean((y_pred - y) * x)
    return dq, dm
```

```python
epochs = 100000
lr = 0.0001
n = len(X)
for epoch in range(epochs): 
    y_pred = q + m * X
    loss = np.mean((y_pred - y)**2)
    dq, dm = derivatives(X, y, y_pred)
    q = q - (lr * dq)
    m = m - (lr * dm)
    if epoch % 10 == 0:
        print(f"Epoca: {epoch} \n Derivata Q: {dq} \n Derivata M: {dm} \n Nuovi parametri: Q: {q}, M: {m} \nLoss {loss}")
```

```python
y_pred = []
x_test = np.array([12,11,4,7,16,35,38,10,18,3,27,5,1,6])
for x in x_test:
    y_pred.append(q + m*x)
y_pred
```

```python
plt.scatter(x_test, y)

plt.plot(x_test, m*x_test + q, color="red")

plt.xlabel("X")
plt.ylabel("Y")
```
